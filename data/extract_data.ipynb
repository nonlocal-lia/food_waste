```python
generated_absolute = table[0]
generated_percent_total = table[1]
recycled = table[2]
composted = table[3]
other_label = table[4]
other_values = table[5]
recycled_percent = table[6]
composted_percent = table[7]
other_label_percent = table[8]
other_values_percent = table[9]
combusted = table[10]
combusted_percent = table[11]
landfilled = table[12]
landfilled_percent = table[13]
```

### Fixing Divided Tables

```python
other_label
```

```python
other_label.iloc[0,0] = other_label.iloc[0,0] + ' ' + other_label.iloc[1,0]
other_label.iloc[2,0] = other_label.iloc[2,0] + ' ' + other_label.iloc[3,0]
other_label.iloc[6,0] = other_label.iloc[6,0] + ' ' + other_label.iloc[7,0]
other_label
```

```python
other_label.drop([1,3,7] ,inplace=True)
other_label.reset_index(inplace=True)
other_label.drop(columns=['index'], inplace=True)
other_label
```

```python
other_label = other_label.columns.to_frame().T.append(other_label, ignore_index=True)
```

```python
other_label
```

```python
other_values = other_values.columns.to_frame().T.append(other_values, ignore_index=True)
other_values
```

```python
other = pd.concat([other_label, other_values], axis=1)
other.columns = ['Material', 'Amount']
```

```python
other
```

```python
other_values_percent = other_values_percent.columns.to_frame().T.append(other_values_percent, ignore_index=True)
other_values_percent
```

```python
other_percent = pd.concat([other_label, other_values_percent], axis=1)
other_percent.columns = ['Material', 'Percent']
other_percent
```

```python
generated_absolute.iloc[3,0] = generated_absolute.iloc[2,0] +" - "+generated_absolute.iloc[3,0]
generated_absolute.iloc[4,0] = generated_absolute.iloc[2,0] +" - "+generated_absolute.iloc[4,0]
generated_absolute.iloc[5,0] = generated_absolute.iloc[2,0] +" - "+generated_absolute.iloc[5,0]
generated_absolute.iloc[6,0] = generated_absolute.iloc[2,0] +" - "+generated_absolute.iloc[6,0]
generated_absolute.drop([2,13], inplace=True)
```

```python
generated_absolute.rename(columns={generated_absolute.columns[0]: 'Material'}, inplace=True)
generated_absolute.set_index('Material', inplace=True)
generated_absolute
```
