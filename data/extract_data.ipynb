### Fixing Divided Tables

Two of the tables didn't load well and the labels and values were stored in seperate tables, we will clean them up so their rows align and then concat them together

```python
other_label
```

```python
# Fixing labels that interpreter treated as seperate lines
other_label.iloc[0,0] = other_label.iloc[0,0] + ' ' + other_label.iloc[1,0]
other_label.iloc[2,0] = other_label.iloc[2,0] + ' ' + other_label.iloc[3,0]
other_label.iloc[6,0] = other_label.iloc[6,0] + ' ' + other_label.iloc[7,0]

# droping extra lines
other_label.drop([1,3,7] ,inplace=True)

# resetting index
other_label.reset_index(inplace=True)
other_label.drop(columns=['index'], inplace=True)
other_label
```

```python
# column title is actually another row label, creating row from column title
other_label = other_label.columns.to_frame().T.append(other_label, ignore_index=True)
other_label
```

```python
# column title is actually a value, creating row from column title
other_values = other_values.columns.to_frame().T.append(other_values, ignore_index=True)
other_values
```

```python
# Joining and relabeling columns
other = pd.concat([other_label, other_values], axis=1)
other.columns = ['Material', 'Amount']
other
```

```python
# column title is actually a value, creating row from column title
other_values_percent = other_values_percent.columns.to_frame().T.append(other_values_percent, ignore_index=True)
other_values_percent
```

```python
# Since labels in both tables are identical, joining percent values with already fixed labels
other_percent = pd.concat([other_label, other_values_percent], axis=1)
other_percent.columns = ['Material', 'Percent']
other_percent
```

## Merging Data

```python
generated_absolute = generated_absolute.T
```

```python
recycled = recycled.T
recycled = recycled.add_prefix('Recycled ')
recycled
```

```python
composted = composted.T
composted = composted.add_prefix('Composted ')
composted
```

```python
combusted = combusted.T
combusted = combusted.add_prefix('Combusted ')
combusted
```

```python
landfilled = landfilled.T
landfilled = landfilled.add_prefix("Landfilled ")
landfilled
```

```python
recycled.reset_index(inplace=True)
recycled.drop(columns=['index'], inplace=True)
recycled.columns.name = None
recycled
```

```python
composted.reset_index(inplace=True)
composted.drop(columns=['index'], inplace=True)
composted.columns.name = None
composted
```

```python
combusted.reset_index(inplace=True)
combusted.drop(columns=['index'], inplace=True)
combusted.columns.name = None
combusted
```

```python
landfilled.reset_index(inplace=True)
landfilled.drop(columns=['index'], inplace=True)
landfilled.columns.name = None
landfilled
```

```python
generated_absolute.columns.name = None
generated_absolute.reset_index(inplace=True)
generated_absolute
```

```python
merged = pd.concat([generated_absolute, recycled, composted, combusted, landfilled], axis=1)
```
